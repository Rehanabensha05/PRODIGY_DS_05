```python
#speed compared to year
plt.figure(figsize=(20,6))
sns.countplot(x='Year',data=accident,hue='Vehicle Speed')
plt.title("Distribution of Speed  ")

plt.show()
```

```python
sns.countplot(x='Year',data=accident,hue='Weather Conditions')
plt.title(" Distribution Weather condition ")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(x='Date',y='Time',data=accident)
plt.title("Accident Based on Date  & Time")
plt.xticks(rotation=60)
plt.show()
```
